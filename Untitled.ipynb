```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

import re
import string
```

```python
# data_fake = pd.read_csv("Fake.csv")
# data_true = pd.read_csv("True.csv")

data_fake = pd.read_csv("Large_Fake_News.csv")
data_true = pd.read_csv("Large_True_News.csv")
```

```python
data_fake_manual_testing["class"] = 0
data_true_manual_testing["class"] = 1
```

```python
data_merge = pd.concat([data_fake, data_true], axis=0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(["title", "subject", "date"], axis=1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)  # shuffle
```

```python
data.head()
```

```python
data.reset_index(inplace=True)
# data.drop(['index', 'level_0'], axis=1, inplace=True)
data.drop(['index'], axis=1, inplace=True)
```

```python
data.columns
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub(r"\[.*?\]", "", text)
    text = re.sub(r"\\W", "", text)
    text = re.sub(r"https?://\S+|www\.\S+", "", text)
    text = re.sub(r"<.*?>+", "", text)
    text = re.sub(r"[%s]" % re.escape(string.punctuation), '', text)
    text = re.sub(r"\n", "", text)
    text = re.sub(r"\w*\d\w*", "", text)
    return text
```

```python
data["text"] = data["text"].apply(wordopt)
```

```python
x = data["text"]
y = data["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
lr = LogisticRegression()
lr.fit(xv_train, y_train)
```

```python
pred_lr = lr.predict(xv_test)
```

```python
lr.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
dt = DecisionTreeClassifier()
dt.fit(xv_train, y_train)
```

```python
pred_dt = dt.predict(xv_test)
```

```python
dt.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
gb = GradientBoostingClassifier(random_state=0)
gb.fit(xv_train, y_train)
```

```python
pred_gb = gb.predict(xv_test)
```

```python
gb.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
rf = RandomForestClassifier(random_state=0)
rf.fit(xv_train, y_train)
```

```python
pred_rf = rf.predict(xv_test)
```

```python
rf.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_label(n):
    if n == 0:
        return "Fake news"
    elif n == 1:
        return "Not a Fake news"

def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_lr = lr.predict(new_xv_test)
    pred_dt = dt.predict(new_xv_test)
    pred_gb = gb.predict(new_xv_test)
    pred_rf = rf.predict(new_xv_test)

    result = {
        "Logistic Regression": output_label(pred_lr[0]),
        "Decison Tree": output_label(pred_dt[0]),
        "Gradient Boosting": output_label(pred_gb[0]),
        "Random Forest": output_label(pred_rf[0])
    }

    return result
```
